```python
from sklearn import neighbors, datasets, preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns
```

```python
df1 = df1.round()
df1.head()
```

```python
df_RM = pd.get_dummies(df1['RM'])
df_RM.head()
```

```python
df_hab = df1.join(df_RM)
df_hab
```

```python
df_hab.groupby(['RM']).mean()
```

```python
# normalización media
def mean_norm(df_input):
    return df_input.apply(lambda x: (x-x.mean())/ x.std(), axis=0)

df_mean_norm = mean_norm(df1)

df_mean_norm = mean_norm(df1)
print(df_mean_norm)
```

```python
sns.histplot(data=df_mean_norm, x='CRIM')
```

```python
sns.histplot(data=df1, x='CRIM')
```

```python
sns.histplot(data=df_mean_norm, x='AGE')
```

```python
# normalizació amb min max
def minmax_norm(df_input):
    return (df1 - df1.min()) / ( df1.max() - df1.min())

df_minmax_norm = minmax_norm(df1)

print(df_minmax_norm)
```

```python
sns.histplot(data=df_minmax_norm, x='CRIM')
```

```python
sns.histplot(data=df_minmax_norm, x='AGE')
```
